```python
content_items = soup.find_all(attrs={"class": "hds-content-item-inner"})

planet_names = []

for item in content_items:
    h3_elements = item.find_all('h3')
    planet_names.extend([h3.text for h3 in h3_elements])

print(planet_names)
```

```python
con_items = soup.find_all(attrs={"class": "CustomField"})

details = []

for item in content_items:
    span_elements = item.find_all('span')
    details.extend([span.text for span in span_elements])

print(details)
```

```python
exoplanets = []

current_exoplanet = {}

for i in range(0, len(details), 2):
    key = details[i]
    value = details[i+1]
    
    if key == 'Light-Years From Earth: ':
        if current_exoplanet:
            exoplanets.append(current_exoplanet)
            current_exoplanet = {}
    
    current_exoplanet[key.strip(': ')] = value

if current_exoplanet:
    exoplanets.append(current_exoplanet)
for exoplanet in exoplanets:
    print(exoplanet)
```

```python
# Printing some page's data as there are total 377 pages which needs high compilation and fetching power to print the data

import requests
from bs4 import BeautifulSoup

base_url = 'https://science.nasa.gov/exoplanets/exoplanet-catalog/?pageno={}&content_list=true'

all_planet_names = []
all_details = []

for page_number in range(1,377):
    url = base_url.format(page_number)
    
    page = requests.get(url)
    
    if page.status_code == 200:
        soup = BeautifulSoup(page.text, 'html.parser')
        
        content_items = soup.find_all(attrs={"class": "hds-content-item-inner"})
        
        for item in content_items:
            h3_elements = item.find_all('h3')
            all_planet_names.extend([h3.text.strip() for h3 in h3_elements])
        
        con_items = soup.find_all(attrs={"class": "CustomField"})
        
        for item in con_items:
            span_elements = item.find_all('span')
            all_details.extend([span.text.strip() for span in span_elements])
    else:
        print(f"Failed to retrieve page {page_number}")



print("completed")



```

```python
import pandas as pd
import numpy as np


df = pd.read_csv(r"C:\Users\divya\Downloads\exoplanet_data.csv")
df.head()
```

```python
df['Planet_Mass'] = df['Planet_Mass'].apply(lambda x: f"{round(float(x.split()[0]) * 1300,3)} Earths" if 'Jupiters' in x else x)
```

```python
def convert_mass(mass):
    if 'Earths' in mass:
        return float(mass.replace(' Earths', ''))
df['Planet_Mass'] = df['Planet_Mass'].apply(convert_mass)
```
